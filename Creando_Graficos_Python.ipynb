#Visualización de datos con Python: Matplotlib, Seaborn y Plotly

```python
import pandas as pd

df = pd.read_csv("inmigrantes_canada.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.set_index("Pais",inplace = True)
```

```python
anos = list(map(str,range(1980,2014)))

anos
```

```python
colombia = df.loc["Colombia",anos]
colombia
```

```python
col_dict = {"Año":colombia.index.tolist(),"Inmigrantes":colombia.values.tolist()}
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col["Año"], datos_col["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
```

```python
plt.figure(figsize=(8,4))
plt.plot(datos_col["Año"], datos_col["Inmigrantes"])
plt.xticks(["1980","1985","1990","1995","2000","2005","2010"])
plt.title("Inmigrantes de Colombia hacia Canada")
plt.xlabel("Años")
plt.ylabel("Inmigrantes")
plt.show()
```

```python
brasil = df.loc["Brasil",anos]
brasil
bra_dict = {"Año":brasil.index.tolist(),"Inmigrantes":brasil.values.tolist()}
bra_dict
datos_bra = pd.DataFrame(bra_dict)
datos_bra
```

```python
argentina = df.loc["Argentina",anos]
argentina
arg_dict = {"Año":argentina.index.tolist(),"Inmigrantes":argentina.values.tolist()}
arg_dict
datos_arg = pd.DataFrame(arg_dict)
datos_arg
```

```python
df_comparacion = df.loc[['Brasil', 'Argentina'], anos]
df_comparacion = df_comparacion.T
df_comparacion.head()
plt.plot(df_comparacion['Brasil'], label='Brasil')
plt.plot(df_comparacion['Argentina'], label='Argentina')
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(datos_col["Año"], datos_col["Inmigrantes"])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_title("Inmigrantes de Colombia hacia Canada\n1980 -2013")
ax.set_xlabel("Años")
ax.set_ylabel("Inmigrantes")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 3))
axs[0].plot(datos_col["Año"], datos_col["Inmigrantes"])
axs[0].set_title("Inmigrantes de Colombia hacia Canada\n1980-2013")
axs[0].set_xlabel("Años")
axs[0].set_ylabel("Inmigrantes")
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
axs[1].boxplot(datos_col["Inmigrantes"])
axs[1].set_title("Boxplot de inmigración de colombianos\nhacia Canada 1980 -2013")
axs[1].set_xlabel("Colombia")
axs[1].set_ylabel("Numero de Inmigrantes")
axs[1].grid()

plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace = 0.3)
fig.suptitle("Inmigración de los cuatro mayores países sudamericanos\n hacia Canadá desde 1980 a 2013", fontsize=12)

axs[0,0].plot(df.loc["Colombia",anos])
axs[0,0].set_title("Colombia")

axs[0,1].plot(df.loc["Brasil",anos])
axs[0,1].set_title("Brasil")

axs[1,0].plot(df.loc["Argentina",anos])
axs[1,0].set_title("Argentina")

axs[1,1].plot(df.loc["Perú",anos])
axs[1,1].set_title("Perú")

ymin = 0
ymax = 7000

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel("Años")
    ax.set_ylabel("Inmigrantes")
    ax.grid()

plt.show()
```

```python
tiendas = ['A', 'B', 'C', 'D']
ventas_2022 = {'Ene': [100, 80, 150, 50],
    'Feb': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'May': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Sep': [240, 160, 290, 130],
    'Oct': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dec': [300, 350, 400, 250]}

df = pd.DataFrame(ventas_2022, index=tiendas)

fig, axs = plt.subplots(2, 2, figsize=(10, 6))
fig.subplots_adjust(hspace=0.5, wspace = 0.3)
fig.suptitle("Ventas de las Tiendas", fontsize=12)

axs[0,0].plot(df.loc["A"])
axs[0,0].set_title("Ventas en la tienda A")

axs[0,1].plot(df.loc["B"])
axs[0,1].set_title("Ventas en la tienda B")

axs[1,0].plot(df.loc["C"])
axs[1,0].set_title("Ventas en la tienda C")

axs[1,1].plot(df.loc["D"])
axs[1,1].set_title("Ventas en la tienda D")

ymin = 0
ymax = 450

for ax in axs.ravel():
  ax.set_ylim(ymin,ymax)


for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel("Mes")
    ax.set_ylabel("Numero de Ventas")
    ax.grid()

```

```python
import pandas as pd

df = pd.read_csv("inmigrantes_canada.csv")
df.head()
```

```python
sudamerica = df.query("Region == 'América del Sur'")
sudamerica.head(10)
```

```python
colores = ["royalblue", "orange", "forestgreen", "orchid", "purple","brown", "slateblue", "gray", "olive", "navy", "teal", "tomato"]

fig, ax = plt.subplots(figsize = (10,5))
ax.bar(sudamerica["Pais"], sudamerica["Total"], color = colores)
ax.set_title("Inmigración de Sudamericanos hacia Canada\n 1980 - 2013", loc = "left", fontsize = 18)
ax.set_xlabel("")
ax.set_ylabel("Número de Inmigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = "--")
plt.show()
```

```python
colores = ["royalblue", "orange", "forestgreen", "orchid", "purple","brown", "slateblue", "gray", "olive", "navy", "teal", "tomato"]

fig, ax = plt.subplots(figsize = (10,5))
ax.barh(sudamerica["Pais"], sudamerica["Total"], color = colores)
ax.set_title("Inmigración de Sudamericanos hacia Canada\n 1980 - 2013", loc = "left", fontsize = 18)
ax.set_ylabel("")
ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = "--")
plt.show()
```

```python
sudamerica_sorted = sudamerica.sort_values(by = "Total", ascending = False)
sudamerica_sorted.head()
```

```python
colores = ["royalblue", "orange", "forestgreen", "orchid", "purple","brown", "slateblue", "gray", "olive", "navy", "teal", "tomato"]

fig, ax = plt.subplots(figsize = (10,3))
ax.barh(sudamerica_sorted["Pais"], sudamerica_sorted["Total"], color = colores)
ax.set_title("Inmigración de Sudamericanos hacia Canada\n 1980 - 2013", loc = "left", fontsize = 18)
ax.set_ylabel("")
ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = "--")
plt.show()
```

```python
colores = []

for index, row in sudamerica_sorted.iterrows():
  if row["Pais"] == "Colombia":
    colores.append("orange")
  else:
    colores.append("silver")


fig, ax = plt.subplots(figsize = (10,3))

ax.set_frame_on(False)
ax.xaxis.set_visible(False)
ax.tick_params(axis = "y", which = "both", bottom = False, top = False, labelbottom = False)

ax.barh(sudamerica_sorted["Pais"], sudamerica_sorted["Total"], color = colores)

ax.barh(sudamerica_sorted["Pais"], sudamerica_sorted["Total"], color = colores)
ax.set_title("Inmigración de Colombianos hacia Canada\n 1980 - 2013", loc = "left", fontsize = 18)
ax.set_ylabel("")
ax.set_xlabel("Número de Inmigrantes", fontsize = 14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
ax.grid(linestyle = "--")

for i,j in enumerate(sudamerica_sorted["Total"]):
  ax.text(j+20, i, str(j), color = "black", fontsize=8.5, ha = "left", va = "center")

plt.show()
```
